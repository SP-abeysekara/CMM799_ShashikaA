```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data/chunk_0_with_info.csv")
df.head()
```

### EDA (Add new column named profit)

```python
df['profit'] = df['revenue'] - df['budget']
df.head()
```

### Filter only the relevant data

```python
# Define the list of columns you want to keep, including production_countries
columns_to_keep = ['id', 'vote_average', 'release_date', 'profit', 'genres', 
                   'production_countries', 'production_companies', 'spoken_languages', 
                   'directors', 'main_characters']

# Select these columns from the original DataFrame
df_selected = df[columns_to_keep]

# Display the first few rows to verify
df_selected.head()
```

## Production Country

```python
df_selected_countries = df_selected.copy()
```

```python
df_selected_countries.head()
```

```python
# Split the 'production_countries' column into individual countries
df_selected_countries['production_countries'] = df_selected['production_countries'].str.split(', ')

# Explode the countries into separate rows
df_exploded = df_selected_countries.explode('production_countries')
df_exploded.head()
```

```python
# Count the occurrences of each country
country_count = df_exploded['production_countries'].value_counts()
print(country_count)
```

```python
# Filter out countries with less than threshold occurrences
threshold = 20
filtered_country_count = country_count[country_count >= threshold]
print(filtered_country_count)
```

```python
# Plot the histogram
filtered_country_count.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Number of Movies Produced by Each Country')
plt.xlabel('Production Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter the original exploded DataFrame to include only the countries with at least 10 movies
df_filtered = df_exploded[df_exploded['production_countries'].isin(filtered_country_count.index)]

# Calculate the average IMDb rating for each country
country_ratings = df_filtered.groupby('production_countries')['vote_average'].mean().sort_values(ascending=False)
print(country_ratings)

# Plotting the average IMDb rating for each country
country_ratings.plot(kind='bar', figsize=(10, 6), color='lightgreen')
plt.title(f'Average IMDb Ratings by Production Country (Count â‰¥ {threshold})')
plt.xlabel('Production Country')
plt.ylabel('Average IMDb Rating')
plt.xticks(rotation=45)
plt.show()
```

```python

# Assuming df_selected is your DataFrame with 'production_countries', 'vote_average', and 'budget'

# Explode the production_countries so each country has its own row
df_exploded = df_selected.explode('production_countries')

# Remove rows where production_countries are NaN (if any)
df_exploded = df_exploded.dropna(subset=['production_countries'])

# Optionally, you can filter out countries with less than a certain number of movies
country_counts = df_exploded['production_countries'].value_counts()
df_exploded = df_exploded[df_exploded['production_countries'].isin(country_counts[country_counts >= 10].index)]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_exploded['production_countries'], df_exploded['vote_average'], alpha=0.5)
plt.title('Production Countries vs Vote Average')
plt.xlabel('Production Countries')
plt.ylabel('Vote Average')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_exploded['production_countries'], df_exploded['profit'], alpha=0.5)
plt.title('Production Countries vs Budget')
plt.xlabel('Production Countries')
plt.ylabel('Budget')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Released Year

```python
df_released_year = df_selected.copy()
```

```python
# Ensure the 'release_date' column is in datetime format
df_released_year['release_date'] = pd.to_datetime(df_released_year['release_date'], errors='coerce')

# Create a new column for the release year
df_released_year['release_year'] = df_released_year['release_date'].dt.year

# Drop rows with NaN in 'release_year' (if any)
df_released_year = df_released_year.dropna(subset=['release_year'])

# Convert 'release_year' to integer (optional, if it's not already)
df_released_year['release_year'] = df_released_year['release_year'].astype(int)

# Calculate the number of movies released each year
movies_per_year = df_released_year['release_year'].value_counts().sort_index()

# Define the threshold value
threshold = 20

# Filter years with movie counts greater than or equal to the threshold
filtered_years = movies_per_year[movies_per_year >= threshold]

# Plot the number of movies released each year, with counts above the threshold
plt.figure(figsize=(12, 7))
filtered_years.plot(kind='bar', color='skyblue')
plt.title('Number of Movies Released Each Year (Count â‰¥ 30)')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()


```

```python
# Calculate the average IMDb rating for each release year
avg_rating_per_year = df_released_year.groupby('release_year')['vote_average'].mean().sort_index()

# Plot the average IMDb rating for each release year
plt.figure(figsize=(12, 7))
avg_rating_per_year.plot(kind='line', marker='o', color='orange')
plt.title('Average IMDb Rating by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average IMDb Rating')
plt.grid(True)
plt.show()

# Calculate total budget for each release year
total_budget_per_year = df_released_year.groupby('release_year')['profit'].sum().sort_index()

# Plot the total budget for each release year
plt.figure(figsize=(12, 7))
total_budget_per_year.plot(kind='line', marker='o', color='green')
plt.title('Profit by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```
